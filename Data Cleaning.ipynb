```python
train['Min_exp'] = train['experience'].apply(lambda x: x.split('-')[0])
test['Min_exp'] = test['experience'].apply(lambda x: x.split('-')[0])
```

```python
train
```

```python
def seniority(des):
    if des <= 3: 
        return 'Entry Level'
    elif des>4 and des<=7:
        return 'Mid Level'
    else:
        return 'Senior Level'
```

```python
train['Min_exp'] = train['Min_exp'].astype('int32')
test['Min_exp'] = test['Min_exp'].astype('int32')
```

```python
train['Seniority'] = train['Min_exp'].apply(seniority)
train['Seniority'].value_counts()
```

```python
test['Seniority'] = test['Min_exp'].apply(seniority)
test['Seniority'].value_counts()
```
