```python
train.job_type.replace('analytics','Analytics',inplace=True)
train.job_type.replace('ANALYTICS','Analytics',inplace=True)
train.job_type.replace('Analytic','Analytics',inplace=True)
train.job_type.replace('analytic','Analytics',inplace=True)

test.job_type.replace('analytics','Analytics',inplace=True)
test.job_type.replace('ANALYTICS','Analytics',inplace=True)
test.job_type.replace('Analytic','Analytics',inplace=True)
test.job_type.replace('analytic','Analytics',inplace=True)
```

```python
def analytics(des):
    if 'analyst' in des.lower() or 'analytics' in des.lower() or 'analysis' in des.lower() or 'scientist' in des.lower() or 'science' in des.lower() or 'machine' in des.lower():
        return 'Analytics'
    else:
        return 'Other'
    

```

```python
train['job_type'] = train['job_desig'].apply(analytics)
```

```python
train['Min_exp'] = train['experience'].apply(lambda x: x.split('-')[0])
test['Min_exp'] = test['experience'].apply(lambda x: x.split('-')[0])
```

```python
def seniority(des):
    if des <= 3: 
        return 'Entry Level'
    elif des>4 and des<=7:
        return 'Mid Level'
    else:
        return 'Senior Level'
```

```python
train['Min_exp'] = train['Min_exp'].astype('int32')
test['Min_exp'] = test['Min_exp'].astype('int32')
```

```python
train['Seniority'] = train['Min_exp'].apply(seniority)
train['Seniority'].value_counts()
```

```python
test['Seniority'] = test['Min_exp'].apply(seniority)
test['Seniority'].value_counts()
```

```python
def city(c):
    if 'delhi' in c.lower() or 'bengaluru' in c.lower() or 'mumbai' in c.lower() or 'hyderabad' in c.lower() or 'pune' in c.lower() or 'gurgaon' in c.lower() or 'chennai' in c.lower() or 'noida' in c.lower():
        return c
    else:
        return 'Other'
```

```python
train['location'] = train['location'].astype('str')
test['location'] = test['location'].astype('str')
```

```python
train['City'] = train['location'].apply(city)
test['City'] = test['location'].apply(city)
```
